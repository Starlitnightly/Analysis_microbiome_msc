```python
import omicverse as ov
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ov.plot_set()
```

# Alpha
## evenness

```python
plot_data=pd.read_csv('qiime_result/alpha_raw/evenness-group-significance-metadata.tsv',sep='\t',index_col=0)
plot_data.head()
```

```python
import scanpy as sc
adata=sc.AnnData(plot_data[['pielou_evenness']])
adata.obs=plot_data.copy()
adata.obs.columns=['Group','key']
#adata=adata[bacteria_group]
adata
```

```python
adata.var
```

```python
adata.uns['Group_colors']=['#EC9DC5','#D4D4D4','#5BC23D','#000000','#8D8DF7']
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ov.pl.bardotplot(adata,groupby='Group',color='pielou_evenness',
                 figsize=(6,2),
           ax=ax,
          ylabel='Faith’s Phylogenetic Diversity',
           bar_kwargs={'alpha':0.5,'linewidth':2,'width':0.6,'capsize':4},
           scatter_kwargs={'alpha':0.8,'s':10,'marker':'o'})
plt.title('Alpha diversity',fontsize=13)
#plt.savefig(f'figures/bar-alpha-faith.png',dpi=300,bbox_inches='tight')
#plt.savefig(f'pdf/bar-alpha-faith.pdf',dpi=300,bbox_inches='tight')
```

```python
import scanpy as sc
fig, ax = plt.subplots(figsize=(4,4))
sc.pl.violin(adata,keys='pielou_evenness',groupby='Group',show=False,ax=ax
            )

plt.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['left'].set_position(('outward', 10))
ax.spines['bottom'].set_position(('outward', 10))

plt.ylabel('Evenness Diversity',fontsize=13)
            # Automatically adjust the plot parameters to give the plot elements some space
ax.set_xticklabels(ax.get_xticklabels(),rotation=90, 
                   horizontalalignment='right',fontsize=12)
plt.title('Alpha diversity',fontsize=13)
plt.ylim(0,1)
plt.tight_layout()
plt.savefig(f'figures/violin-alpha-evenness.png',dpi=300,bbox_inches='tight')
plt.savefig(f'pdf/violin-alpha-evenness.pdf',dpi=300,bbox_inches='tight')
```

# Beta



## jaccard_distance_matrix

```python
plot_data2=pd.read_csv('qiime_result/beta_raw/jaccard_pcoa_results/ordination.txt',
                       sep='\t',header=None,index_col=0)
plot_data2.head()
```

```python
adata.obsm['jaccard_pcoa']=plot_data2.loc[adata.obs.index].values
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ov.pl.embedding(adata,basis='jaccard_pcoa',
               color='Group',size=500,
                frameon='small',show=False,ax=ax)
plt.title('Beta Diversity',fontsize=13)
plt.savefig(f'figures/pcoa-jaccard.png',dpi=300,bbox_inches='tight')
plt.savefig(f'pdf/pcoa-jaccard.pdf',dpi=300,bbox_inches='tight')
```

```python
distance=pd.read_csv('qiime_result/beta_raw/jaccard_distance_matrix/distance-matrix.tsv',sep='\t',index_col=0)
distance.head()
```

```python
fig, ax = plt.subplots(figsize=(6,6))
sns.heatmap(distance,square=True,cmap='RdBu',
            ax=ax,cbar_kws={'shrink':0.8},yticklabels=True,xticklabels=True)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90, 
                   horizontalalignment='right',fontsize=12)
ax.set_yticklabels(ax.get_yticklabels(),
                   horizontalalignment='right',fontsize=12)
plt.title('Beta Jaccard Distance',fontsize=13)
plt.savefig(f'figures/pcoa-jaccard-heatmap.png',dpi=300,bbox_inches='tight')
plt.savefig(f'pdf/pcoa-jaccard-heatmap.pdf',dpi=300,bbox_inches='tight')
```
